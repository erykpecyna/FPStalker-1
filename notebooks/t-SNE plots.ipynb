```python
import numpy as np
from sklearn.manifold import TSNE
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import pandas as pd
X = np.load("../X_sample.npy")
y = np.load("../y_sample.npy")
```

```python
tsne = TSNE(n_components=3, random_state=0)
X_tsne = tsne.fit_transform(X)
```

```python
from mpl_toolkits.mplot3d import Axes3D

tsne = TSNE(n_components=3, random_state=0)
X_tsne = tsne.fit_transform(X)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

groups = pd.DataFrame(X_tsne, columns=['x', 'y', 'z']).assign(category=y).groupby('category')
for name, points in groups:
    ax.scatter(points.x, points.y, points.z, label=name)

ax.legend()
```

```python
tsne = TSNE(n_components=2, random_state=0)
X_tsne = tsne.fit_transform(X)
fig = plt.figure()
ax = fig.add_subplot(111)

groups = pd.DataFrame(X_tsne, columns=['x', 'y']).assign(category=y).groupby('category')
for name, points in groups:
    ax.scatter(points.x, points.y, label=name)

ax.legend()
```
